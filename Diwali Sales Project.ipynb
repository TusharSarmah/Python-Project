```python
# import python libraries
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
# import csv file
df=pd.read_csv("Diwali Sales Data.csv", encoding="Unicode_Escape")
```

```python
df.shape
```

```python
df.head(10)
```

```python
df.info()
```

```python
#drop unrelated/blank columns
df.drop(["Status","unnamed1"],axis=1,inplace= True)
```

```python
df.shape
```

```python
#check for null values
df.isnull()
```

```python
#check for null values
pd.isnull(df).sum()
```

```python
#Removing Null Values
df.dropna(inplace=True)
```

```python
#Rechecking Null Values 
pd.isnull(df).sum()
```

```python
#Changing Amount type to Integer
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtype
```

```python
#describe() function gives us additional info of the data (mean,std,max,min,etc)
df.describe()
```

```python
#Using descirbe function to get additional data of particular data
df[["Age","Orders","Amount"]].describe()

```

```python
# plotting a bar chart for Gender and it's count
ax=sns.countplot (x='Gender',data=df, hue="Gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Table for Gender Data vs Amount
df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by='Amount', ascending=False)
```

```python
# plotting a bar chart for gender vs total amount
sales_gen=df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen,hue="Gender")


```

### From The above analysis we found that most of the buyers are female and has more purchasing power than men.


#### AGE

```python
#Using bar graph to show the number of buyers in different age groups
ax=sns.countplot(data=df, x="Age Group", hue="Gender")
sns.set(rc={'figure.figsize':(5,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Bar Graph to show total amount spend by different age groups
sales_age=df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Age Group",y="Amount",data=sales_age,hue="Age Group")
```

### From the above graphs we could find that most of the buyers are from 25-35 age group where most of the buyers are female

### State

```python
# total number of orders from top 10 states
sales_state=df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending= False).head(10)
sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(data=sales_state,x="State",y="Orders",hue="State")
```

```python
# total amount/sales from top 10 states
sales_state=df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False,).head(10)
sns.set(rc={"figure.figsize":(17,7)})
sns.barplot(data=sales_state,x="State",y="Amount",hue="State")
```

### From the above graphs we could find that most of the orders and sales are from Uttar Pradesh, Maharashtra and Karnataka

### Marital Status

```python
#Bar graph showing married vs unmarried count
ax=sns.countplot(x="Marital_Status",data=df,hue="Marital_Status")
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Bar graph showing purchasing power of male and female on the basis of marital status
sales_Marital=df.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.set(rc={"figure.figsize":(6,6)})
sns.barplot(x="Marital_Status",y="Amount",hue="Gender",data=sales_Marital)
```

### From the above bar graphs we could find that married female buyers have high purchasing power

### Occupation

```python
#Bar graph showing count of buyers from various industry
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(x="Occupation",data=df,hue="Occupation")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Bar graph showing purchasing power of buyers from various industry
Sales_Occupation=df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Occupation",y="Amount",data=Sales_Occupation,hue="Occupation")
```

### From the above bar graph we could analyze that IT,Healthcare and Aviation has the highest purchasing power

### Product Category

```python
# Bar graph showing count of buyers in various prouct category
ax=sns.countplot(data=df,x="Product_Category",hue="Product_Category")
sns.set(rc={"figure.figsize":(25,8)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Bar graph showing sales in different Product category
Sales_Product=df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(16)
sns.set(rc={"figure.figsize":(30,10)})
sns.barplot(data=Sales_Product,x="Product_Category",y="Amount",hue="Product_Category")
```

### From the above bar graphs we could find that most of the buyers are from food, clothing & Apparel, Electronics & Gadgets and Footwear & Shoes. As well as the number of buyers is higher in Clothing & Apparel category but the revenue or amount generated is higher in Food category.

### Product

```python
# Bar graph showing top 10 products sold
Sales_ProductID=df.groupby(["Product_ID"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
sns.barplot(data=Sales_ProductID,x="Product_ID",y="Orders",hue="Product_ID")
```

### Conclusion

### Married women age group 26-35 yrs from UP,  Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category
